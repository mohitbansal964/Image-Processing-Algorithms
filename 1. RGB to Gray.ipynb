# 1. RGB to Gray Scale Image

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
from PIL import Image
import cv2
import numpy as np
```

```python
arr= cv2.imread("./Images/car.jpg")
arr= cv2.cvtColor(arr,cv2.COLOR_BGR2RGB)
arr.shape
```

```python
plt.imshow(arr)
plt.title('Original Image')
plt.show()
```

```python
red= arr.copy()
red[:, :, 1]= 0
red[:, :, 2]= 0

green= arr.copy()
green[:, :, 0]= 0
green[:, :, 2]= 0

blue= arr.copy()
blue[:, :, 0]= 0
blue[:, :, 1]= 0

gray= np.dot(arr, [0.299, 0.587, 0.114])
```

```python
fig, ax= plt.subplots(ncols= 3, figsize= (18, 6))
fsize= 14

ax[0].set_title("Red Plane", fontsize= fsize)
ax[0].imshow(red)

ax[1].set_title("Green Plane", fontsize= fsize)
ax[1].imshow(green)

ax[2].set_title("Blue Plane", fontsize= fsize)
ax[2].imshow(blue)

plt.show()
```

```python
plt.title('Gray Scale Image')
plt.imshow(gray, cmap= "gray")
plt.show()
```

```python
img.imsave("./Images/car_gray.png", gray, cmap="gray")
```
